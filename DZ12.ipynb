Можно использовать Matplotlib, Seaborn или сохранить результаты в

```python
import pyshark
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


file_path = 'dns-remoteshell.pcap'


capture = pyshark.FileCapture(file_path)


packet_count = 0
http_packets = 0
tcp_packets = 0
udp_packets = 0

# Итерация по пакетам
for packet in capture:
    packet_count += 1

    # Проверка на наличие слоев
    if 'http' in packet:
        http_packets += 1
    if 'tcp' in packet:
        tcp_packets += 1
    if 'udp' in packet:
        udp_packets += 1

    # Вывод информации о пакете (опционально, для детального анализа)
    # print(packet)

# Закрытие захвата
capture.close()

data = {
    'Protocol': ['Total', 'HTTP', 'TCP', 'UDP'],
    'Count': [packet_count, http_packets, tcp_packets, udp_packets]
}
df = pd.DataFrame(data)


sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x='Protocol', y='Count', data=df, palette='viridis')
plt.title('Анализ пакетов в файле pcapng')
plt.xlabel('Протокол')
plt.ylabel('Количество пакетов')
plt.show()
```
